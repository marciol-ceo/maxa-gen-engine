```python

import requests
import json

BASE_URL = "http://localhost:5000"

# ============================================
# TEST 1 : G√©n√©ration d'UN SEUL exercice
# ============================================
def test_single_exercise():
    """G√©n√®re un seul exercice al√©atoire"""
    url = f"{BASE_URL}/generate/exercise/random"
    
    payload = {
        "index_name": "gen-engine-index",
        "n_variations_per_exercice": 3,
        "temperature": 0.8,
        "return_all_latex": False
    }
    
    response = requests.post(url, json=payload)
    
    if response.status_code == 200:
        result = response.json()
        print("‚úÖ Exercice g√©n√©r√© avec succ√®s!")
        print(f"Source chunk ID: {result['source_chunk_id']}")
        print(f"\nLaTeX g√©n√©r√©:\n{result['latex_result'][:500]}...")  # Affiche les 500 premiers caract√®res
        
        # Sauvegarder dans un fichier
        with open("exercice_test.tex", "w", encoding="utf-8") as f:
            f.write(result['latex_result'])
        print("\nüìÑ Fichier sauvegard√©: exercice_test.tex")
    else:
        print(f"‚ùå Erreur {response.status_code}: {response.text}")

# ============================================
# TEST 2 : G√©n√©ration d'une √âPREUVE COMPL√àTE (mode mixed)
# ============================================
def test_full_epreuve_mixed():
    """G√©n√®re une √©preuve compl√®te avec exercices de tous les namespaces"""
    url = f"{BASE_URL}/generate/auto"
    
    payload = {
        "index_name": "gen-engine-index",
        "mode": "mixed",  # Prend des exercices de CHAQUE namespace
        "n_variations_per_exercice": 5,
        "temperature": 0.7,
        "return_all_latex": True
    }
    
    response = requests.post(url, json=payload)
    
    if response.status_code == 200:
        result = response.json()
        print("‚úÖ √âpreuve compl√®te g√©n√©r√©e (mode MIXED)!")
        print(f"Mode utilis√©: {result['mode_used']}")
        print(f"Nombre de chunks: {result['chunks_count']}")
        print(f"\nLaTeX g√©n√©r√©:\n{result['latex_result'][:500]}...")
        
        # Sauvegarder
        with open("epreuve_mixed_test.tex", "w", encoding="utf-8") as f:
            f.write(result['latex_result'])
        print("\nüìÑ Fichier sauvegard√©: epreuve_mixed_test.tex")
    else:
        print(f"‚ùå Erreur {response.status_code}: {response.text}")

```
